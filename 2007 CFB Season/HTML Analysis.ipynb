```python
#I realized it was easier to use read_html to get the table directly

#url for website
url='https://www.sports-reference.com/cfb/schools/'
```

```python
#create list of dataframes (only 1)
list_of_df = pd.read_html(url)
```

```python
#assign first dataframe
df=list_of_df[0]
```

```python
print(df.shape)
df.head()
```

```python
#remove hierarchy and replace headers
df.columns = ['Rank', 'School', 'From', 'To', 'Years', 'Overall Games','Overall Wins', 'Overall Losses', 'Overall Ties',
              'Overall Win Pct', 'Bowl Games', 'Bowl Wins', 'Bowl Losses', 'Bowl Ties', 'Bowl Win Pct', 'SRS', 
              'Strength of Schedule', 'AP Polls', 'Conference Championships', 'Notes']
df.head()
```

```python
#remove first row as it provides no value
df.drop(columns=['Rank'], inplace=True)
print(df.shape)
df.head()
```

```python
#drop rows with header labels (occur every 20 entries)
drop_rows=[]


for i in range(326):
    if i==20:
        drop_rows.append(i)
    elif i==21:
        drop_rows.append(i)
    elif (i+2)%22==0:
        drop_rows.append(i)
        drop_rows.append(i+1)
    else:
        pass
        
df.drop(drop_rows, inplace=True)      
```

```python
#format columns to appropriate data types
df['From']=df['From'].astype(int)
df['To']=df['To'].astype(int)
df['Years']=df['Years'].astype(int)
df['Overall Games']=df['Overall Games'].astype(int)
df['Overall Wins']=df['Overall Wins'].astype(int)
df['Overall Losses']=df['Overall Losses'].astype(int)
df['Overall Ties']=df['Overall Ties'].astype(int)
df['Overall Win Pct']=df['Overall Win Pct'].astype(float)
df['Bowl Games']=df['Bowl Games'].astype(int)
df['Bowl Wins']=df['Bowl Wins'].astype(int)
df['Bowl Losses']=df['Bowl Losses'].astype(int)
df['Bowl Ties']=df['Bowl Ties'].astype(int)
df['Bowl Win Pct']=df['Bowl Win Pct'].astype(float)
df['SRS']=df['SRS'].astype(float)
df['Strength of Schedule']=df['Strength of Schedule'].astype(float)
df['AP Polls']=df['AP Polls'].astype(int)
df['Conference Championships']=df['Conference Championships'].astype(int)
```

```python
#drop schools that did not play in 2007 and later as that is the relevant year for overall project
df[df['To']>=2007]
```

```python
#find z scores for years
years_z_score = stats.zscore(df['Years'])

#find z scores for overall games
games_z_score = stats.zscore(df['Overall Games'])

#find z scores for overall wins
wins_z_score = stats.zscore(df['Overall Wins'])

#find z scores for bowl games
bowl_games_z_score = stats.zscore(df['Bowl Games'])

#find z scores for bowl wins
bowl_wins_z_score = stats.zscore(df['Bowl Wins'])

#find z scores for conference championships
conference_z_score = stats.zscore(df['Conference Championships'])
```

```python
#no outliers outside 3 standard deviations
#check for outliers outside 2 standard deviations for years
df[(years_z_score<-2)|(years_z_score>2)]
```

All entries within 2 standard deviations

```python
#no outliers outside 3 standard deviations
#check for outliers outside 2 standard deviations for games
df[(games_z_score<-2)|(games_z_score>2)]
```

All entries within 2 standard deviations

```python
#no outliers outside 3 standard deviations
#check for outliers outside 2 standard deviations for wins
#sort by descending
df[(wins_z_score<-2)|(wins_z_score>2)].sort_values(by=['Overall Wins'], ascending=False)
```

Most of these outliers are "blue blood" programs along with some other strong programs.

```python
#check for outliers outside 3 standard deviations for bowl games
df[(bowl_games_z_score<-3)|(bowl_games_z_score>3)].sort_values(by=['Bowl Games'], ascending=False)
```

Again, these outliers for bowl games contain the blue bloods Alabama, Texas, and USC. Georgia is another historic program.

```python
#check for outliers outside 3 standard deviations for bowl wins
df[(bowl_wins_z_score<-3)|(bowl_wins_z_score>3)].sort_values(by=['Bowl Wins'], ascending=False)
```

Again, mostly blue bloods or other very successful programs.

```python
#check for outliers outside 3 standard deviations for conference championships
df[(conference_z_score<-3)|(conference_z_score>3)].sort_values(by=['Conference Championships'], ascending=False)
```

Yet again, all blue bloods...and Tulsa.
